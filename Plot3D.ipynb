```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import torch
from torch import nn
```

```
def plot_3d(weights, bias):
  x1 = np.linspace(-1,1,10)
  x2 = np.linspace(-1,1,10)
  X1,X2 = np.meshgrid(x1, x2)

  # Transformação linear em 3D
  # w1*x1 + w2*x2 + w3*x3 + b = 0
  # Isolando x3 #
  # x3 = (-b - w1*x1 - w2*x2)/w3

  X3 = (bias[0] - weights[0]*X1 - weights[1]*X2) / weights[2]

  fig = plt.figure(figsize=(10, 8))
  ax = fig.gca(projection='3d')
  ax.view_init(azim=180)
  
  surf = ax.plot_surface(X1, X2, X3, cmap='plasma')
  plt.xlabel('X')
  plt.ylabel('Y')
```

```
linear = nn.Linear(in_features=3, out_features=1)
print(linear)

weights = linear.weight.data.numpy().squeeze()
bias    = linear.bias.data.numpy()

plot_3d(weights, bias)
```
